```python
import numpy as np
import biomart as biomart
```

```python
def merge_genespectra_low_spec_cl_ensembl(species_name_file:str, species_id:str, species_name: str):
    print(species_name_file)
    species_classes = pd.read_csv(f"data/1TPM_lim_taxo_species/{species_name_file}_classes_taxo_cs_1TPM_ensembl_name.csv")
    species_classes_lowspec = preprocess_genespectra_output_low_spec(species_classes)

    species_classes_lowspec['tissue_name'] = cell_ontology_no_ct.loc[cell_ontology_no_ct.species_scientific_name == species_name]['tissue_name'].values[0]
    species_classes_lowspec['uberon_tissue_id'] = cell_ontology_no_ct.loc[cell_ontology_no_ct.species_scientific_name == species_name]['uberon_tissue_id'].values[0]
    species_classes_lowspec['ncbi_txid'] = cell_ontology_no_ct.loc[cell_ontology_no_ct.species_scientific_name == species_name]['ncbi_txid'].values[0]
    species_classes_lowspec['species_scientific_name'] = cell_ontology_no_ct.loc[cell_ontology_no_ct.species_scientific_name == species_name]['species_scientific_name'].values[0]


    print(species_classes_lowspec.shape)
    
    server = biomart.BiomartServer("http://www.ensembl.org/biomart")
    dataset = server.datasets[f'{species_name}_gene_ensembl']

    attributes = ['ensembl_gene_id', 'external_gene_name']
    filters = {'external_gene_name': np.unique(species_classes_lowspec['external_gene_name'].values.astype(str))}  

    print("Downloading ensembl response")

    response = dataset.search({'query': filters,
            'attributes': attributes,
            'mart_instance': 'ensembl'})

    gene_to_id = pd.read_csv(response.url, sep='\t', header=None, names=attributes)
    gene_to_id = gene_to_id.dropna()

    species_classes_final = species_classes_lowspec.merge(gene_to_id, left_on='external_gene_name', right_on='external_gene_name', how='left').dropna().reset_index(drop=True)

# same columns as specific genespectra file
    
    species_classes_final['specificity_category_type'] = 'low cell type specificity'
    species_classes_final['cell_type_name'] = 'NA'
    species_classes_final['cell_ontology_id'] = 'NA'


    print(species_classes_final.shape)

    species_classes_final.to_csv(f"data/1TPM_lim_taxo_species/{species_name_file}_classes_taxo_cs_1TPM_ensembl_name_lowspec_processed_all.csv", index=False)

    
    # return species_classes_final
```

```python
merge_genespectra_low_spec_cl_ensembl(species_name_file='human', species_id='9606', species_name='hsapiens')
```

```python
merge_genespectra_low_spec_cl_ensembl(species_name_file='gorilla', species_id='9593', species_name='ggorilla')
merge_genespectra_low_spec_cl_ensembl(species_name_file='chimp', species_id='9598',  species_name='ptroglodytes')
merge_genespectra_low_spec_cl_ensembl(species_name_file='macaque', species_id='9544', species_name='mmulatta')
merge_genespectra_low_spec_cl_ensembl(species_name_file='marmoset', species_id='9483', species_name='cjacchus')
```
